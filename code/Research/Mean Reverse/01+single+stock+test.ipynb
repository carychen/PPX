```python
apple_price_this_year = get_pricing(24, start_date='2017-01-01', end_date='2017-05-25')["price"]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apple_price_this_year

apple_price_mvag_5d = pd.rolling_mean(apple_price_this_year, window=5).rename('mvag_past_5d')
apple_price_mvag_10d = pd.rolling_mean(apple_price_this_year, window=10).rename('mvag_past_10d')
apple_price_mvag_20d = pd.rolling_mean(apple_price_this_year, window=20).rename('mvag_past_20d')
apple_price_mvag_30d = pd.rolling_mean(apple_price_this_year, window=30).rename('mvag_past_30d')
apple_price_mvag_60d = pd.rolling_mean(apple_price_this_year, window=60).rename('mvag_past_60d')

apple_price_this_year_mvag = pd.concat([apple_price_this_year, 
            apple_price_mvag_5d, 
          apple_price_mvag_10d,
          apple_price_mvag_20d,
          apple_price_mvag_30d,
          apple_price_mvag_60d], axis=1
         )
apple_price_this_year_mvag
```

```python
plt.plot(apple_price_this_year.index, apple_price_this_year.values)
plt.plot(apple_price_mvag_30d.index, apple_price_mvag_30d.values)
```

```python
apple_price_1d_return = apple_price_this_year.pct_change(-1).rename('return_next_1d')
apple_price_2d_return = apple_price_this_year.pct_change(-2).rename('return_next_2d')
apple_price_3d_return = apple_price_this_year.pct_change(-3).rename('return_next_3d')
apple_price_4d_return = apple_price_this_year.pct_change(-4).rename('return_next_4d')
apple_price_5d_return = apple_price_this_year.pct_change(-5).rename('return_next_5d')
apple_price_10d_return = apple_price_this_year.pct_change(-10).rename('return_next_10d')
apple_price_20d_return = apple_price_this_year.pct_change(-20).rename('return_next_20d')
apple_price_30d_return = apple_price_this_year.pct_change(-30).rename('return_next_30d')
apple_price_60d_return = apple_price_this_year.pct_change(-60).rename('return_next_60d')

apple_price_this_year_return = pd.concat([apple_price_1d_return, 
                                        apple_price_2d_return,
                                        apple_price_3d_return,
                                        apple_price_4d_return,
                                        apple_price_5d_return,
                                        apple_price_10d_return,
                                        apple_price_20d_return,
                                        apple_price_30d_return,
                                        apple_price_60d_return])

```
